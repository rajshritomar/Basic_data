## Q.1 Prepare dataset

```python
import pandas as pd
import numpy as np
```

```python
file_data = 'C:/Users/hp/OneDrive/Desktop/python_1st/Python Foundation Case Study 1 - Retail Case Study/'
```

```python
Customer = pd.read_csv(file_data + 'Customer.csv')
Customer
```

```python
Customer.info()
```

```python
Transaction = pd.read_csv(file_data + 'Transactions.csv')
Transaction
```

```python
Transaction.info()
```

```python
Product_Heirarchy = pd.read_csv(file_data + 'prod_cat_info.csv')
Product_Heirarchy
```

```python
## merging
customer_trans = pd.merge(left = Customer , right = Transaction , left_on = 'customer_Id' , right_on = 'cust_id' ,
                         how = 'inner' , indicator = True)
customer_trans
```

```python
Customer_final = pd.merge(left = customer_trans , right = Product_Heirarchy  , left_on = 'prod_cat_code' , right_on = 'prod_cat_code' ,
                         how = 'inner' )
Customer_final
```

```python
## converting dob and trans_date from object to date
Customer_final['DOB'] = pd.to_datetime( Customer_final['DOB'] , format = '%d-%m-%Y')
Customer_final['DOB'].head()
```

```python
Customer_final['tran_date'] = pd.to_datetime( Customer_final['tran_date'] , format = '%d-%m-%Y' )
Customer_final['tran_date'].head()
```

```python
Customer_final.duplicated().sum()
```

```python
Customer_final.drop_duplicates(inplace = True)
```

## Q.5 - Analyze which product categories are more popular among females vs male customers.

```python
## popular among males
M = Customer_final.loc[(Customer_final.Gender == 'M')].groupby(['prod_cat'])['Qty'].sum().nlargest(3)
print('Most popular prod cat in males :' , M)

## popular among females
F = Customer_final.loc[(Customer_final.Gender == 'F')].groupby(['prod_cat'])['Qty'].sum().nlargest(3)
print('Most popular prod cat in females :' , F)
```
